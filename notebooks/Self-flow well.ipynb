```python
def trace(data_x, data_y, namexy):
    tracep = go.Scattergl(
        x = data_x,
        y = data_y,
        name = namexy,
        mode = 'lines'
    )
    return tracep

def plot():
    layout = dict(title = 'Расчет фонтанирующей скважины',  yaxis=dict(title = 'Измеренная глубина, м', autorange='reversed'))

    fig = dict(data=data, layout=layout)

    iplot(fig, filename='basic-scatter')
```

```python
trace1 = trace(simple_well.data.get_values(27), simple_well.data.get_values(26),  "Давление, бар")
trace2 = trace(simple_well.data.get_values(28), simple_well.data.get_values(26),  "Температура флюида, С")
trace3 = trace(simple_well.data.get_values(29), simple_well.data.get_values(26),  "Температура земли, С")
data=[trace1,trace2,trace3]
plot()
```
