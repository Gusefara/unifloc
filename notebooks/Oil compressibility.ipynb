## Коэффициент сжимаемости нефти $c_o$

Коэффициент сжимаемости характеризует уменьшение объема нефти при увеличении давления

Также в литературе можно встретить тождественное название - коэффициент объемной упругости

Корреляция при давлении выше давления насыщения была получена с помощью модели линейной регрессии на 4,036 замеренный точках

$$ p \geq p_b $$

$$c_o = \frac{ a_1 + a_2 R_s + a_3 T + a_4\gamma_{gs} + a_5 \gamma_o  }{ a_6 p}$$
где:

$a_1 = -1433.0$,

$a_2 = 5.0$,

$a_3 = 17.2$,

$a_4 = - 1180.0$,

$a_5 = 12.61$,

$a_6 = 10^5$,
