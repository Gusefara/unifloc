```python
import math
import matplotlib.pyplot as plt
r_omkm=1.15
lenth_km=2
u_kv=3.3
i_A=22
c_cuprum=385
rho_kgm3=8930
A=22*10**(-6)
volume_m3=A*lenth_km*1000
m=volume_m3*rho_kgm3
def p(i,r,l):
    return i**2*r*l
def delta(q,c,m):
    return q/c/m
```

```python
time=[]
tfromheat=[]
tfromheat2=[]
for t in range(1000):
    time.append(t/60)
    power=p(i_A,r_omkm,lenth_km)
    deltat=delta(power*t,c_cuprum,m)
    tfromheat.append(deltat)
    power=p(40,r_omkm,lenth_km)
    deltat=delta(power*t,c_cuprum,m)
    tfromheat2.append(deltat)
    
    
    
```

```python
plt.plot(time,tfromheat,label='Изменение температуры от времени')
plt.plot(time,tfromheat2)
plt.xlabel('время, минут')
plt.ylabel('Изменение температуры')
plt.show()   
```
